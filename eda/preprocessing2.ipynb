```python
def show_tags(movie):
    '''
    Takes in a string of a movie title, prints the tags for that movie.
    '''
    movie_idx = movie_df.loc[movie_df.title == movie].index[0]
    movie_tags = movie_df.tags.loc[movie_idx]
    for tag in movie_tags:
        print tag
    return movie_tags
```

```python
# Way to search for the exact movie title in the dataframe if you don't know it
def search_title(search_string):
    result_df = movie_df.loc[movie_df.title.str.contains(search_string)]
    return result_df
```
